```python
def get_data(df, column_name, model_name):
    data_enrich = df[column_name].describe()
    data_enrich['mode'] = df[column_name].mode()[0].item()
    data_enrich['model_name'] = model_name
    return data_enrich

result = get_data(enrich_dbscan, 'enrichment_factor1', 'dbscan local')
result = pd.concat([result, get_data(enrich_kmeans, 'enrichment_factor1', 'kmeans local')], axis=1)
result = pd.concat([result, get_data(enrich_kproto, 'enrichment_factor1', 'kprototypes local')], axis=1)
result = pd.concat([result, get_data(enrich_dbscan, 'enrichment_factor2', 'dbscan global')], axis=1)
result = pd.concat([result, get_data(enrich_kmeans, 'enrichment_factor2', 'kmeans global')], axis=1)
result = pd.concat([result, get_data(enrich_kproto, 'enrichment_factor2', 'kprototypes global')], axis=1)

result = pd.concat([result, get_data(df_atac, 'Length', 'ATAC')], axis=1)
result = pd.concat([result, get_data(df_dnase, 'Length', 'DNase')], axis=1)

print(result)
result.to_csv('../outputs/final/stats.csv', sep='\t')
```
