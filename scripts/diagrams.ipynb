```python

import pyreadr
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import numpy as np
```

```python
df_kmeans = pd.read_csv("../outputs/final/ranges2_kmeans.csv", sep='\t')
```

```python
bins = np.arange(0, 1001, step=50)

df_kmeans['Length'] = df_kmeans['End'] - df_kmeans['Start']
plt.xlim(0, 1000)
plt.ylim(0, 5000)
plt.xticks(bins, fontsize = 10, rotation='vertical')
plt.title("TOP-seq k-means")
plt.xlabel('Regiono ilgis')
plt.ylabel('Dažnis')

sns.histplot(data=df_kmeans, bins=bins, x="Length")
```
