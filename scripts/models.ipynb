```python
def calculate_gaps(pyranges_obj):
    """
    Calculate gaps between ranges in a PyRanges object.

    Parameters:
        pyranges_obj (pr.PyRanges): A PyRanges object with genomic ranges.

    Returns:
        pr.PyRanges: A PyRanges object containing the gaps.
    """
    # Convert PyRanges to a DataFrame
    df = pyranges_obj.df
    
    # Sort the ranges by Chromosome and Start position
    df = df.sort_values(["Chromosome", "Start"]).reset_index(drop=True)
    
    # Shift End and Start columns to calculate gaps
    prev_end = df.groupby("Chromosome")["End"].shift(0)   # Current End
    next_start = df.groupby("Chromosome")["Start"].shift(-1)  # Next Start
    
    # Identify gaps where End < Next Start
    gaps_df = df[(prev_end < next_start)].assign(Start=prev_end, End=next_start)[["Chromosome", "Start", "End"]]
    
    # Drop NaN rows caused by shifting and ensure integer type
    gaps_df = gaps_df.dropna().astype({"Start": "int", "End": "int"})
    
    # Convert back to PyRanges
    return pr.PyRanges(gaps_df)
```
